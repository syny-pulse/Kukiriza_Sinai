```python
# Load necessary libraries
import os
os.environ["OMP_NUM_THREADS"] = "1"
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# load the dataset
df = pd.read_csv('customer_data_records.csv')
```

```python
# Select Features for clustering
x = df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
# Apply K-Means clustering
# Use only 2 clusters for simplicity
kmeans = KMeans(n_clusters = 3, random_state=0)
kmeans.fit(x)
```

```python
# Ad   Cluster to dataframe
df['Cluster'] = kmeans.labels_
df.head(10)
```

```python
# visulalize the cluster
plt.figure(figsize=(9,7))
sns.scatterplot(data = df, x = 'Annual Income (k$)', y = 'Spending Score (1-100)',hue = 'Cluster')
plt.title('Customer clusters based on income and spending')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.show()
```
